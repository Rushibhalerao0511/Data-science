# Data visualization

Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.


# 1.Histograms:
Plot histograms for each numerical column.
Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.


```python
import matplotlib.pyplot as plt
```

```python
# Plot histograms for each numerical column
for column in df.select_dtypes(include='number'):
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
# Analyze the distribution
skewness = df[column].skew()
outliers = df[(df[column] < df[column].quantile(0.25) - 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25))) |
                  (df[column] > df[column].quantile(0.75) + 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25)))]
    
if outliers.empty:
    outlier_status = "No outliers"
else:
    outlier_status = "Outliers present"

print(f"Analysis for column '{column}':")
print(f"Skewness: {skewness}")
print(outlier_status)
print()
```

# 2.Boxplots:



```python
# Plot boxplots for each numerical column
plt.figure(figsize=(10, 6))
df.boxplot()
plt.title('Boxplots of Numerical Variables')
plt.xticks(rotation=45)
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
# Identify outliers and discuss findings
for column in df.select_dtypes(include='number'):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    print(f"Analysis for column '{column}':")
    print(f"Interquartile Range (IQR): {iqr}")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    
    if outliers.empty:
        print("No outliers")
    else:
        print("Outliers:")
        print(outliers)
    
    print()
```

# Standardization of Numerical Variables
Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.


#step1
Concept of standardization (z-score normalization).

Standardization, also known as z-score normalization, is a technique used to transform numerical data into a standard scale with a mean of 0 and a standard deviation of 1. This transformation helps to make different variables comparable by removing the units of measurement and scaling the data to have a consistent spread.

The formula for standardization (z-score normalization) is:

𝑧=𝑥-mu/sigma

Where:

z= the standardized value.

x= the original value.

mu= the mean of the variable.

sigma= the standard deviation of the variable.

```python
# Before standardization: Plot histograms for each numerical column
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns):
    plt.subplot(2, len(numerical_columns)//2, i+1)
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Before Standardization: {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Standardize numerical columns
for column in numerical_columns:
    mu = df[column].mean()
    sigma = df[column].std()
    df[column + '_standardized'] = (df[column] - mu) / sigma
```

```python
# After standardization: Plot histograms for standardized numerical columns
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns):
    standardized_column = column + '_standardized'
    plt.subplot(2, len(numerical_columns)//2, i+1)
    plt.hist(df[standardized_column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'After Standardization: {standardized_column}')
    plt.xlabel(standardized_column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

# Conversion of Categorical Data into Dummy Variables
Objective: To transform categorical variables into a format that can be provided to ML algorithms.


#Need for Converting Categorical Data into Dummy Variables:

Algorithm Compatibility: Most machine learning algorithms require numerical input data. Categorical variables, which contain non-numeric values like strings or categories, cannot be directly used as input to these algorithms.

Preservation of Information: Converting categorical variables into numerical representations while preserving the information they carry is essential. One-hot encoding achieves this by creating binary (0 or 1) columns for each category, indicating the presence or absence of that category in the original variable.

```python
#Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.

# Select categorical columns to one-hot encode
categorical_columns = df.select_dtypes(include='object').columns

# Apply one-hot encoding to categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_columns)

# Display a portion of the transformed dataset
print("Transformed Dataset with Dummy Variables:")
df_encoded.head()
```

Importance of Data Preprocessing Steps:

Enhanced Model Performance: Data preprocessing steps like standardization and one-hot encoding play a crucial role in improving the performance of machine learning models. Standardization helps algorithms converge faster and prevents features with larger scales from dominating the learning process. One-hot encoding ensures that categorical variables are appropriately represented in a numerical format, preventing bias and ensuring accurate model predictions.

Improved Interpretability: Standardization and one-hot encoding contribute to the interpretability of machine learning models by making the input features more understandable and interpretable. Standardized features are on a consistent scale, allowing for easier comparison and interpretation of their effects on the model's predictions. One-hot encoded variables provide clear representations of categorical information, facilitating insights into the impact of different categories on the model's output.

Data Integrity and Consistency: Preprocessing steps ensure the integrity and consistency of the data used for analysis and modeling. Standardization and one-hot encoding help remove inconsistencies, such as varying scales or incompatible data types, ensuring that the input data is suitable for analysis and model training.
