### 1.Data Cleaning and Preparation:



```python
import pandas as pd
```

```python
#Load the dataset into a DataFrame or equivalent data structure.
df=pd.read_csv('Cardiotocographic.csv')
df.head(2125)
```

```python
# Handle missing values (e.g. imputation, deletion)

data_cleaned1=df.drop(['DS','DP'],axis=1) #deleting column 
data_cleaned1

data_cleaned1[data_cleaned1.isnull().any(axis=1)].head()
```

```python
# Impute missing values with the mean
data_cleaned1.fillna(data_cleaned1.mean(), inplace=True)
data_cleaned1
```

```python
#Detect and treat outliers if necessary.

data_cleaned1.boxplot()
```

```python
# finding outlier using z-score
from scipy.stats import zscore

# Define the threshold for Z-score
threshold = 3

# Calculate Z-score for each numerical column
z_scores = data_cleaned1.apply(zscore)

# Identify outliers using Z-score
outliers = (z_scores > threshold) | (z_scores < -threshold)
outliers

#emove rows containing outliers
data_cleaned1 = data_cleaned1[~outliers.any(axis=1)]
```

Provide a statistical summary for each variable in the dataset, including measures of central tendency (mean, median) and dispersion (standard deviation, interquartile range).

### Data Visualization:


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Select numerical columns for visualization
numerical_columns = ['LB', 'AC', 'FM', 'UC', 'DL', 'ASTV', 'MSTV', 'ALTV', 'MLTV']

# Create histograms to visualize the distributions of numerical variables
plt.figure(figsize=(12, 8))
for column in numerical_columns:
    plt.hist(data_cleaned1[column], bins=20, alpha=0.5, label=column)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histograms of Numerical Variables')
plt.legend()
plt.show()


# Create boxplots to visualize the distributions of numerical variables
plt.figure(figsize=(12, 8))
sns.boxplot(data=data_cleaned1[numerical_columns])
plt.xlabel('Numerical Variables')
plt.ylabel('Values')
plt.title('Boxplots of Numerical Variables')
plt.xticks(rotation=45)
plt.show()


# Create scatter plots to explore relationships between pairs of numerical variables
plt.figure(figsize=(12, 8))
sns.pairplot(data_cleaned1[numerical_columns])
plt.title('Pair Plot of Numerical Variables')
plt.show()


# Create correlation heatmap to explore relationships between pairs of numerical variables
plt.figure(figsize=(10, 8))
sns.heatmap(data_cleaned1[numerical_columns].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()


# Create violin plots for deeper insights into the distribution of numerical variables
plt.figure(figsize=(12, 8))
for column in numerical_columns:
    sns.violinplot(x=column, data=data_cleaned1, inner='quartile')
plt.title('Violin Plots of Numerical Variables')
plt.show()
```
